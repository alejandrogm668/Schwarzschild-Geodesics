```python
import numpy as np 
import matplotlib.pyplot as plt
from scipy import * 
from scipy.integrate import odeint
plt.rcParams["figure.figsize"] = (15,10)
plt.rcParams.update({'font.size': 22})
```

```python
Î¼ = 1
c = 1.5

def F_Ï(r,Ï,Î¼=Î¼,c=c):
    return -np.sqrt(2*Î¼*c**2/r)

Ï = np.linspace(0,10,1000)
r0 = 10

sol_Ï = odeint(F_Ï,r0,Ï)
r_Ï = sol_Ï.T[0]

def F_t(r,t,Î¼=Î¼,c=c):
    return -np.sqrt(2*Î¼*c**2)*(1-2*Î¼/r)/np.sqrt(r)

t = np.linspace(0,20,1000)

sol_t = odeint(F_t,r0,t)
r_t = sol_t.T[0]

fig, axs = plt.subplots(2,1)
axs[0].plot(Ï, r_Ï)
axs[0].set_xlabel(r"$\tau$")
axs[0].set_ylabel(r"$r$")
axs[1].plot(t,r_t)
axs[1].set_xlabel(r"$t$")
axs[1].set_ylabel(r"$r$")
fig.tight_layout()
plt.savefig('2_radial_motion.png', format='png')
```
